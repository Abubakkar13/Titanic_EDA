```python
titanic_df.isnull().sum()
```

```python
#After confirming with emptyness, will fill those with the mean of that coloumn
titanic_df.fillna(titanic_df.mean(), inplace=True)
titanic_df.isnull().sum()
```

### Now will drop/separate a column with more than 50% of empty cells.

```python
z = titanic_df.isnull().sum()
drop_col = z[z > (50/100 * titanic_df.shape[0])]
drop_col
```

Therefore, after the execution of above code snippet we come to know that <b>Cabin</b> column is the one with more that 50% of empty cells

```python
drop_col.index
```

```python
titanic_df.drop(drop_col.index, axis=1, inplace=True)
titanic_df.isnull().sum()
```

```python
#Now we cannot calculate mean for the String values,
#hence we go for the one whose frequency of occurance is maximum to fill the empty cell
titanic_df['Embarked'].describe()
```

```python
titanic_df['Embarked'].fillna('S', inplace=True)
titanic_df.isnull().sum()
```

And Finally all the empty cells have been filled with respective known value

# <center>Data Visualization with EDA</center>

### Few randomly created visualization on Titanic survival Dataset
Now, to use Matplotlib, Seaborn library to visualize the Dataset.

```python
plt.figure(figsize=(10,6))
sns.countplot('Survived',data=titanic_df)
plt.title('Plot to show the Survived or Not')
plt.show()
```

```python
#Plot to visualize number of Males and Females

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
gender = ['Male','Female']
index = [577,314]
ax.bar(gender,index)
plt.xlabel("Gender")
plt.ylabel("No of people onboarding ship")
plt.show()
```

```python
alive = len(titanic_df[titanic_df['Survived'] == 1])
dead = len(titanic_df[titanic_df['Survived'] == 0])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
status = ['Survived','Dead']
ind = [alive,dead]
ax.bar(status,ind)
plt.xlabel("Status")
plt.show()
```

```python
titanic_df.groupby('Sex')[['Survived']].mean()
```

```python
plt.figure(1)
titanic_df.loc[titanic_df['Survived'] == 1, 'Pclass'].value_counts().sort_index().plot.bar()
plt.title('Bar graph of people accrding to ticket class in which people survived')


plt.figure(2)
titanic_df.loc[titanic_df['Survived'] == 0, 'Pclass'].value_counts().sort_index().plot.bar()
plt.title('Bar graph of people accrding to ticket class in which people couldn\'t survive')
```

```python
# Now using seaborn to plot the count of passengers in each class

fg = sns.factorplot('Pclass', data=titanic_df, kind='count', aspect=1.5)
fg.set_xlabels('Class')
```

```python
# Number of men and women in each of the passenger class
titanic_df.groupby(['Sex', 'Pclass'])['Sex'].count()
```

```python
# Again use saeborn to group by Sex and class
g = sns.factorplot('Pclass', data=titanic_df, hue='Sex', kind='count', aspect=1.75)
g.set_xlabels('Class')
```

```python
# Drop missing values for the records in which age passenger is missing
age = titanic_df['Age'].dropna()

age_dist = sns.distplot(age)
age_dist.set_title("Distribution of Passengers' Ages")
```
